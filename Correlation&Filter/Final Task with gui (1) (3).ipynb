```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog, messagebox,simpledialog
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import math
import os

# Function to read the signal from a file
def ReadSignalFile(file_name):
    expected_indices=[]
    expected_samples=[]
    with open(file_name, 'r') as f:
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        while line:
            # process line
            L=line.strip()
            if len(L.split(' '))==2:
                L=line.split(' ')
                V1=int(L[0])
                V2=float(L[1])
                expected_indices.append(V1)
                expected_samples.append(V2)
                line = f.readline()
            else:
                break
    return expected_indices,expected_samples

def on_display_click():
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if file_path:
        try:
            # Read the signal from the selected file
            indices, values = ReadSignalFile(file_path)
            # Visualize the signal inside the Tkinter window
            visualize_continuos_signal(Signal_Frame, indices, values)
        except Exception as e:
            messagebox.showerror("Error", f"Could not read file: {e}")
    else:
        messagebox.showinfo("Information", "No file selected.")

def save_signal_to_file(filename, indices, values):
    with open(filename, 'w') as file:
        file.write(f"{len(indices)}\n")  # Write the number of samples
        for index, value in zip(indices, values):
            file.write(f"{index} {value}\n")  # Write each index and value

# Function to visualize the discrete signal using a stem plot inside Tkinter window
def visualize_continuos_signal(window, indices, values):
    # Clear previous plot (if any)
    for widget in window.winfo_children():
        widget.destroy()
    
    # Create the figure for the plot
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(indices, values)
    
    ax.set_xlabel("Sample Index")
    ax.set_ylabel("Signal Value")
    ax.set_title("Continuos Signal Visualization")
    ax.grid(True)
    
    # Embed the figure in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=window)  # A canvas widget for matplotlib
    canvas.draw()
    
    # Position the canvas in the window
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

def visualize_discrete_signal(window, indices, values):
    # Clear previous plot (if any)
    for widget in window.winfo_children():
        widget.destroy()
    
    # Create the figure for the plot
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.stem(indices, values)
    
    ax.set_xlabel("Sample Index")
    ax.set_ylabel("Signal Value")
    ax.set_title("Discrete Signal Visualization")
    ax.grid(True)
    
    # Embed the figure in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=window)  # A canvas widget for matplotlib
    canvas.draw()
    
    # Position the canvas in the window
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

def AddSignalSamplesAreEqual(userFirstSignal,userSecondSignal,Your_indices,Your_samples):
    if(userFirstSignal=='Signal1.txt' and userSecondSignal=='Signal2.txt'):
        file_name=r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\add.txt"  # write here the path of the add output file
    expected_indices,expected_samples=ReadSignalFile(file_name)          
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Addition Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Addition Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Addition Test case failed, your signal have different values from the expected one") 
            return
    print("Addition Test case passed successfully")

def SubSignalSamplesAreEqual(userFirstSignal,userSecondSignal,Your_indices,Your_samples):
    if(userFirstSignal=='Signal1.txt' and userSecondSignal=='Signal2.txt'):
        file_name=r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\subtract.txt" # write here the path of the subtract output file
        
    expected_indices,expected_samples=ReadSignalFile(file_name)   
    
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Subtraction Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Subtraction Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Subtraction Test case failed, your signal have different values from the expected one") 
            return
    print("Subtraction Test case passed successfully")

def Folding(Your_indices,Your_samples):
    file_name = r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\folding.txt"  # write here the path of the folding output file
    expected_indices,expected_samples=ReadSignalFile(file_name)      
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Folding Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Folding Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Folding Test case failed, your signal have different values from the expected one") 
            return
    print("Folding Test case passed successfully")

def ShiftSignalByConst(Shift_value,Your_indices,Your_samples):
    if(Shift_value==3):  #x(n+k)
        file_name = r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\advance3.txt"
    elif(Shift_value==-3): #x(n-k)
        file_name = r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\delay3.txt"
        
    expected_indices,expected_samples=ReadSignalFile(file_name)      
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Shift by "+str(Shift_value)+" Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Shift by "+str(Shift_value)+" Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Shift by "+str(Shift_value)+" Test case failed, your signal have different values from the expected one") 
            return
    print("Shift by "+str(Shift_value)+" Test case passed successfully")

def MultiplySignalByConst(User_Const,Your_indices,Your_samples):
    if(User_Const==5):
        file_name=r"C:\Users\DELL\OneDrive\Desktop\Task 1 testcases and testing functions\mul5.txt"  # write here the path of the mul5 output file
        
    expected_indices,expected_samples=ReadSignalFile(file_name)      
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Multiply by "+str(User_Const)+ " Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Multiply by "+str(User_Const)+" Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Multiply by "+str(User_Const)+" Test case failed, your signal have different values from the expected one") 
            return
    print("Multiply by "+str(User_Const)+" Test case passed successfully")

def on_add_signals_click(): 
    
    file_paths = filedialog.askopenfilenames(
        title="Select Signal Files", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if not file_paths:
        messagebox.showinfo("Information", "No files selected.")
        return

    
    signals = [ReadSignalFile(file) for file in file_paths]
    result_indices, result_values = add_signals(signals)
    AddSignalSamplesAreEqual("Signal1.txt", "Signal2.txt",result_indices,result_values) # call this function with your computed indicies and samples
    visualize_continuos_signal(Signal_Frame, result_indices, result_values)

    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")

def add_signals(signal_data_list):

    if not signal_data_list:
        return [], []

    # the first signal's indices and values
    added_indices = signal_data_list[0][0][:]
    added_values = signal_data_list[0][1][:]

    temp_signal = {}  # Initialize an empty dictionary
    for index, value in zip(added_indices, added_values):
        temp_signal[index] = value  # Assign each index to its corresponding value

    # Iterate through the rest of the signals
    for indices, values in signal_data_list[1:]:
        #for each signal 
        for index, value in zip(indices, values):
            if index in temp_signal:  # Check if the index exists in temp_signal
                temp_signal[index] += value  
            

        # Update added_indices and added_values
        added_indices = list(temp_signal.keys())
        added_values = [temp_signal[index] for index in added_indices]

    # Sort the final added signal by index
    sorted_signal = sorted(zip(added_indices, added_values))
    return zip(*sorted_signal) 

def on_sub_signals_click():
    # Open a file dialog to select multiple signal files
    file_paths = filedialog.askopenfilenames(
        title="Select Signal Files", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if not file_paths:
        messagebox.showinfo("Information", "No files selected.")
        return

    # Read and add the signals
    signals = [ReadSignalFile(file) for file in file_paths]
    result_indices, result_values = sub_signals(signals)
    SubSignalSamplesAreEqual("Signal1.txt", "Signal2.txt",result_indices,result_values)  # call this function with your computed indicies and samples
    visualize_continuos_signal(Signal_Frame, result_indices, result_values)
    

    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")

def sub_signals(signal_data_list):

    if not signal_data_list:
        return [], []

    # the first signal's indices and values
    added_indices = signal_data_list[0][0][:]
    added_values = signal_data_list[0][1][:]

    temp_signal = {}  # Initialize an empty dictionary
    for index, value in zip(added_indices, added_values):
        temp_signal[index] = value  # Assign each index to its corresponding value

    # Iterate through the rest of the signals
    for indices, values in signal_data_list[1:]:
        #for each signal 
        for index, value in zip(indices, values):
            if index in temp_signal:  # Check if the index exists in temp_signal
                temp_signal[index] -= value  
            

        # Update added_indices and added_values
        added_indices = list(temp_signal.keys())
        added_values = [temp_signal[index] for index in added_indices]

    # Sort the final added signal by index
    sorted_signal = sorted(zip(added_indices, added_values))
    return zip(*sorted_signal) 

def on_rev_signals_click():
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return

    # Read and add the signals
    indices, values = ReadSignalFile(file_path)
    result_indices, result_values = rev_signal(indices, values)
    Folding(result_indices, result_values)
    visualize_continuos_signal(Signal_Frame, result_indices, result_values)

    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")   
        
def rev_signal(indices, values):
    # Create a dictionary from indices and values
    signal_dict = dict(zip(indices, values))
    temp_signal = {} 
    
    # Iterate over the original signal_dict
    for key, value in signal_dict.items():  
            temp_signal[-key] = signal_dict[key]  


    # Sort the new signal by indices
    sorted_signal = sorted(zip(list(temp_signal.keys()), list(temp_signal.values())))
    
    return zip(*sorted_signal) 

def shift_signal(indices, values, k):
    # Create a dictionary to map original indices to values
    signal_dict = dict(zip(indices, values))
    
    temp_signal = {}  # Initialize an empty dictionary to store the shifted signal
    
    # Iterate over the original signal_dict
    for key, value in signal_dict.items():  
        temp_signal[key - k] = value  

    
    # Sort the shifted signal by the new indices
    sorted_signal = sorted(temp_signal.items())

    
    new_indices, new_values = zip(*sorted_signal)
    

    ShiftSignalByConst(k, new_indices, new_values)
    

    return list(new_indices), list(new_values)

def on_shift_signal_click():
    # Open a file dialog to select the signal file
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    # Read the signal from the file
    indices, values = ReadSignalFile(file_path)
    # Ask for the shift value (k)
    k = simpledialog.askinteger("Input", "Enter a value for shifting (k steps):")
    if k is None:  # If the user cancels the input
        messagebox.showinfo("Information", "No shift value entered.")
        return
    # Shift the signal by k steps
    result_indices, result_values = shift_signal(indices, values, k)
    visualize_continuos_signal(Signal_Frame, result_indices, result_values)

    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")   
        
def multiply_by_constant (indices,values ,multiplier):    
    # Create a dictionary from indices and values
    new_values = [value * multiplier for value in values]
    MultiplySignalByConst(multiplier,indices, new_values)# call this function with your computed indicies and samples

    return indices, new_values
        
def On_Multiply_Click():
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return

    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 
    # multiplier = get_number()
    multiplier = simpledialog.askfloat("Input", "Enter a multiplier to apply to the signal values:")
    
    if multiplier is None:  # If the user cancels the input
        messagebox.showinfo("Information", "No multiplier entered.")
        return
    result_indices, result_values = multiply_by_constant(indices, values,multiplier)
    visualize_continuos_signal(Signal_Frame, result_indices, result_values)

    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")     
    

######################################################################## Task2

def on_display_discrete():
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if file_path:
        try:
            # Read the signal from the selected file
            indices, values = ReadSignalFile(file_path)
            # Visualize the signal inside the Tkinter window
            visualize_discrete_signal(Signal_Frame2, indices, values)
        except Exception as e:
            messagebox.showerror("Error", f"Could not read file: {e}")
    else:
        messagebox.showinfo("Information", "No file selected.")

def on_display_continuos():
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    
    if file_path:
        try:
            # Read the signal from the selected file
            indices, values = ReadSignalFile(file_path)
            # Visualize the signal inside the Tkinter window
            visualize_continuos_signal(Signal_Frame2, indices, values)
        except Exception as e:
            messagebox.showerror("Error", f"Could not read file: {e}")
    else:
        messagebox.showinfo("Information", "No file selected.")        

def on_cos_signal_click():
    # Create a new window (dialog) for input
    input_window = tk.Toplevel()
    input_window.title("Enter Signal Parameters")
    
    # Create labels and entry widgets for each input
    tk.Label(input_window, text="Amplitude:").grid(row=0, column=0)
    amplitude = tk.Entry(input_window)
    amplitude.grid(row=0, column=1)
    
    tk.Label(input_window, text="Phase Shift (Theta):").grid(row=1, column=0)
    phase_shift = tk.Entry(input_window)
    phase_shift.grid(row=1, column=1)
    
    tk.Label(input_window, text="Analog Frequency:").grid(row=2, column=0)
    analog_frequency = tk.Entry(input_window)
    analog_frequency.grid(row=2, column=1)

    tk.Label(input_window, text="Sampling Frequency:").grid(row=3, column=0)
    sampling_frequency = tk.Entry(input_window)
    sampling_frequency.grid(row=3, column=1)
    
    
    def on_submit():
        try:
            # Retrieve and convert input values
            amplitude_a = float(amplitude.get())
            theta = float(phase_shift.get())
            frequency = float(analog_frequency.get())
            fs = float(sampling_frequency.get())
            
            draw_cos_signal(amplitude_a, theta, frequency,fs)
            
            input_window.destroy()  # Close the input window after submission
            
        except ValueError:
            messagebox.showerror("Invalid input", "Please enter valid numbers.")
    
    # Submit button
    submit_button = tk.Button(input_window, text="plot signal", command=on_submit)
    submit_button.grid(row=4, column=1, pady=10)   

def on_sin_signal_click():
    # Create a new window (dialog) for input
    input_window = tk.Toplevel()
    input_window.title("Enter Signal Parameters")
    
    # Create labels and entry widgets for each input
    tk.Label(input_window, text="Amplitude:").grid(row=0, column=0)
    amplitude = tk.Entry(input_window)
    amplitude.grid(row=0, column=1)
    
    tk.Label(input_window, text="Phase Shift (Theta):").grid(row=1, column=0)
    phase_shift = tk.Entry(input_window)
    phase_shift.grid(row=1, column=1)
    
    tk.Label(input_window, text="Analog Frequency:").grid(row=2, column=0)
    analog_frequency = tk.Entry(input_window)
    analog_frequency.grid(row=2, column=1)

    tk.Label(input_window, text="Sampling Frequency:").grid(row=3, column=0)
    sampling_frequency = tk.Entry(input_window)
    sampling_frequency.grid(row=3, column=1)
    
    
    def on_submit():
        try:
            # Retrieve and convert input values
            amplitude_a = float(amplitude.get())
            theta = float(phase_shift.get())
            frequency = float(analog_frequency.get())
            fs = float(sampling_frequency.get())
            
            draw_sin_signal(amplitude_a, theta, frequency,fs)
            
            input_window.destroy()  # Close the input window after submission
            
        except ValueError:
            messagebox.showerror("Invalid input", "Please enter valid numbers.")
    
    # Submit button
    submit_button = tk.Button(input_window, text="plot signal", command=on_submit)
    submit_button.grid(row=4, column=1, pady=10)  

def draw_cos_signal(amplitude, theta, frequency,fs):
    # Generate time values from 0 to 1 second, assuming a sampling rate of 1000 samples/second

    if fs < 2 * frequency :
        messagebox.showinfo("Information", "there is ailising") 
        
    t_samp = np.linspace(0, 1, int(fs))  #samp
    
    t=np.linspace(0, 1, 1000)   #cont
    
    signal_samp = amplitude * np.cos(2 * np.pi * frequency * t_samp + theta)
    
    signal = amplitude * np.cos(2 * np.pi * frequency * t + theta)

    # Clear previous plot (if any)
    for widget in Signal_Frame2.winfo_children():
        widget.destroy()

    # Create the figure for the plot
    fig, (ax,ax2) = plt.subplots(1,2,figsize=(12, 4))
    
    ax.stem(t_samp, signal_samp)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Amplitude")
    ax.set_title("Cosine Signal with sampling")
    ax.grid(True)

    ax2.plot(t, signal)
    ax2.set_xlabel("Time (s)")
    ax2.set_ylabel("Amplitude")
    ax2.set_title("Cosine Signal")
    ax2.grid(True)
    
    # Embed the figure in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=Signal_Frame2)  
    canvas.draw()
    
    # Position the canvas in the window
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

def draw_sin_signal(amplitude, theta, frequency,fs):
    # Generate time values from 0 to 1 second, assuming a sampling rate of 1000 samples/second

    if fs < 2 * frequency :
        messagebox.showinfo("Information", "there is ailising") 
        
    t_samp = np.linspace(0, 1, int(fs))  #samp
    
    t=np.linspace(0, 1, 1000)   #cont
    
    signal_samp = amplitude * np.sin(2 * np.pi * frequency * t_samp + theta)
    
    signal = amplitude * np.sin(2 * np.pi * frequency * t + theta)

    # Clear previous plot (if any)
    for widget in Signal_Frame2.winfo_children():
        widget.destroy()

    # Create the figure for the plot
    fig, (ax,ax2) = plt.subplots(1,2,figsize=(12, 4))
    
    ax.stem(t_samp, signal_samp) 
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Amplitude")
    ax.set_title("Sin Signal with sampling")
    ax.grid(True)

    ax2.plot(t, signal)
    ax2.set_xlabel("Time (s)")
    ax2.set_ylabel("Amplitude")
    ax2.set_title("Sin Signal")
    ax2.grid(True)
    
    # Embed the figure in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=Signal_Frame2)  
    canvas.draw()
    
    # Position the canvas in the window
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

######################################################################## Task3

def QuantizationTest1(file_name,Your_EncodedValues,Your_QuantizedValues):
    expectedEncodedValues=[]
    expectedQuantizedValues=[]
    with open(file_name, 'r') as f:
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        while line:
            # process line
            L=line.strip()
            if len(L.split(' '))==2:
                L=line.split(' ')
                V2=str(L[0])
                V3=float(L[1])
                expectedEncodedValues.append(V2)
                expectedQuantizedValues.append(V3)
                line = f.readline()
            else:
                break
    if( (len(Your_EncodedValues)!=len(expectedEncodedValues)) or (len(Your_QuantizedValues)!=len(expectedQuantizedValues))):
        print("QuantizationTest1 Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_EncodedValues)):
        if(Your_EncodedValues[i]!=expectedEncodedValues[i]):
            print("QuantizationTest1 Test case failed, your EncodedValues have different EncodedValues from the expected one") 
            return
    for i in range(len(expectedQuantizedValues)):
        if abs(Your_QuantizedValues[i] - expectedQuantizedValues[i]) < 0.01:
            continue
        else:
            print("QuantizationTest1 Test case failed, your QuantizedValues have different values from the expected one") 
            return
    print("QuantizationTest1 Test case passed successfully")

def QuantizationTest2(file_name,Your_IntervalIndices,Your_EncodedValues,Your_QuantizedValues,Your_SampledError):
    expectedIntervalIndices=[]
    expectedEncodedValues=[]
    expectedQuantizedValues=[]
    expectedSampledError=[]
    with open(file_name, 'r') as f:
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        while line:
            # process line
            L=line.strip()
            if len(L.split(' '))==4:
                L=line.split(' ')
                V1=int(L[0])
                V2=str(L[1])
                V3=float(L[2])
                V4=float(L[3])
                expectedIntervalIndices.append(V1)
                expectedEncodedValues.append(V2)
                expectedQuantizedValues.append(V3)
                expectedSampledError.append(V4)
                line = f.readline()
            else:
                break
    if(len(Your_IntervalIndices)!=len(expectedIntervalIndices)
     or len(Your_EncodedValues)!=len(expectedEncodedValues)
      or len(Your_QuantizedValues)!=len(expectedQuantizedValues)
      or len(Your_SampledError)!=len(expectedSampledError)):
        print("QuantizationTest2 Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_IntervalIndices)):
        if(Your_IntervalIndices[i]!=expectedIntervalIndices[i]):
            print("QuantizationTest2 Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(Your_EncodedValues)):
        if(Your_EncodedValues[i]!=expectedEncodedValues[i]):
            print("QuantizationTest2 Test case failed, your EncodedValues have different EncodedValues from the expected one") 
            return
        
    for i in range(len(expectedQuantizedValues)):
        if abs(Your_QuantizedValues[i] - expectedQuantizedValues[i]) < 0.01:
            continue
        else:
            print("QuantizationTest2 Test case failed, your QuantizedValues have different values from the expected one") 
            return
    for i in range(len(expectedSampledError)):
        if abs(Your_SampledError[i] - expectedSampledError[i]) < 0.01:
            continue
        else:
            print("QuantizationTest2 Test case failed, your SampledError have different values from the expected one") 
            return
    print("QuantizationTest2 Test case passed successfully")

def on_quant_signal():

    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    

    input_window = tk.Toplevel()
    input_window.title("Enter Signal Parameters")
    
    # Create labels and entry widgets for each input
    tk.Label(input_window, text="Enter Number").grid(row=0, column=0)
    num = tk.Entry(input_window)
    num.grid(row=0, column=1)
    
    
    selected_option = tk.StringVar(value="bits")

    tk.Radiobutton(input_window, text="bits", variable=selected_option, value="bits", font=("Arial", 14)).grid(row=1, column=1, padx=5, pady=5)
    tk.Radiobutton(input_window, text=" levels ", variable=selected_option, value="levels", font=("Arial", 14)).grid(row=2, column=1, padx=5, pady=5)
   

    def confirm():
                
        num_value = int(num.get())  # Convert entry to an integer
        option = selected_option.get()  # Get selected option

        if option == "levels":
            quantization(num_value, indices, values)
        else:
            lev = 2 ** num_value
            print(lev)
            quantization(lev, indices, values)


    # Button to confirm selection
    tk.Button(input_window, text="Confirm", command=confirm).grid(row=3, column=1)

def quantization(levels_num, indices, values):
    max_value = max(values)
    min_value = min(values)
    delta = (max_value - min_value) / levels_num

    
    intervals = []
    midpoints = []
    current = min_value
    sampled_error=[]
    assigned_interval=[]
    
    for i in range(levels_num):
        interval_start = round(current, 2)
        interval_end = round(current + delta, 2)
        intervals.append((interval_start, interval_end))
        
        midpoint = round((interval_start + interval_end) / 2, 2)
        midpoints.append(midpoint)

        # Update current to the start of the next interval
        current += delta
        
    # Convert indices to binary representation
    num_bits = math.ceil(math.log(levels_num, 2))  # Ensure num_bits is an integer
    binary_indices = []
    new_values2 = []
    for value in values:
        # Find the interval index for the value
        for i, (start, end) in enumerate(intervals):
            if start <= value <= end:
                binary_index = format(i, f'0{num_bits}b')  # Convert index to binary with specified bits
                binary_indices.append(binary_index)
                new_values2.append(midpoints[i])
                sampled_error.append(round(midpoints[i]-value, 3))
                assigned_interval.append(i+1)
                break
        else:
            binary_indices.append('0' * num_bits)  # Default if not found

    total_sum_error=sum(x ** 2 for x in sampled_error)
    count_error=len(sampled_error)
    average_error=total_sum_error/count_error
    print("Intervals:", intervals)
    print("Binary Indices:", binary_indices)
    print("New Values:", new_values2)
    print("sampled_error:", sampled_error)
    print("assigned_interval", assigned_interval)
    print("Average Error", average_error)

    # Assuming QuantizationTest1 is defined elsewhere
    if(levels_num==4):
            QuantizationTest2(r'D:\AAA\Level 4\Semester 7\DSP\Assignments\task3\Quan2_Out.txt',assigned_interval , binary_indices, new_values2,sampled_error)
    else:

            QuantizationTest1(r'D:\AAA\Level 4\Semester 7\DSP\Assignments\task3\Quan1_Out.txt', binary_indices, new_values2)
    return


########################################################################## Task4
def SignalSamplesAreEqual(Your_indices,Your_samples):
    
    file_name=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Tast conv\testcases\testcases\Derivative testcases\2nd_derivative_out.txt"  # write here the path of the add output file
    expected_indices,expected_samples=ReadSignalFile(file_name)          
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Signals are not equal, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Signals are not equal, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Signals are not equal, your signal have different values from the expected one") 
            return
    print("Signals are equal Test case passed successfully")

def move_signal(win_size,indices, values):
    win_size=int(win_size)
    print(len(values) - win_size + 1)
    
    avg_values = []  

    # Compute moving averages
    for i in range(len(values) - win_size + 1):
        avg_values.append(sum(values[i:i + win_size]) / win_size)
    

    avg_indices = indices[:len(avg_values)]

    
    return avg_indices, avg_values

def on_move_signal():
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    win_size = simpledialog.askfloat("Input", "Enter window size:")
    
    if win_size is None:  # If the user cancels the input
        messagebox.showinfo("Information", "No size entered.")
        return
    
    result_indices, result_values=move_signal(win_size,indices, values)

    SignalSamplesAreEqual(result_indices,result_values)
    
    visualize_discrete_signal(Signal_Frame4, result_indices,result_values)
    # Save the result to a new file
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")

def first_D(indices, values):
    # First derivative: Y(n) = x(n) - x(n-1)
    first_derivative = []  
    for i in range(1, len(values)):  
        first_derivative.append(values[i] - values[i-1])

    first_indices = indices[:len(first_derivative)]

    return first_indices,first_derivative

def second_D(indices, values):
    # Second derivative: Y(n) = x(n+1) - 2x(n) + x(n-1)
    second_derivative = []  
    for i in range(1, len(values) - 1):  
        second_derivative.append(values[i+1] - 2*values[i] + values[i-1])

    second_indices = indices[:len(second_derivative)] 

    return second_indices,second_derivative

def on_first_D():
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    result_indices, result_values=first_D(indices, values)
    
    for index, value in zip(result_indices, result_values):
        print(f"({index}, {value})")

    SignalSamplesAreEqual(result_indices, result_values)

    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")

def on_second_D():
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    result_indices, result_values=second_D(indices, values)

    for index, value in zip(result_indices, result_values):
        print(f"({index}, {value})")

    SignalSamplesAreEqual(result_indices, result_values)

    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, result_indices, result_values)
        messagebox.showinfo("Success", "Result saved successfully!")

def manual_convolution_with_signals(indices1, samples1, indices2, samples2):
    result_len = len(samples1) + len(samples2) - 1            #9
    result_samples = np.zeros(result_len)  
    
    for i in range(result_len):
        for j in range(len(samples2)):
            if 0 <= i - j < len(samples1):
                result_samples[i] += samples1[i - j] * samples2[j]
    
    result_indices_start = indices1[0] + indices2[0]
    result_indices = [result_indices_start + i for i in range(result_len)]

    result_samples = np.round(result_samples, 6)
    
    return result_indices, result_samples

def conv_signal_successs(Your_indices,Your_samples):
    
    file_name=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Tast conv\testcases\testcases\Convolution testcases\Conv_output.txt"  # write here the path of the add output file
    expected_indices,expected_samples=ReadSignalFile(file_name)          
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Signals are not equal, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Signals are not equal, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Signals are not equal, your signal have different values from the expected one") 
            return
    print("Signals are equal Test case passed successfully")

def on_conv_signal():

    file1 = filedialog.askopenfilename(title="Select First Signal File")
    if not file1:
        print("No file selected for Signal 1.")
        return

    indices1, samples1 = ReadSignalFile(file1)
    
    file2 = filedialog.askopenfilename(title="Select Second Signal File")
    if not file2:
        print("No file selected for Signal 2.")
        return

    indices2, samples2 = ReadSignalFile(file2)

    result_indices, result_samples = manual_convolution_with_signals(indices1, samples1, indices2, samples2)
    
    conv_signal_successs(result_indices,result_samples)
    
    print("Convolved Signal:")
    print(f"Length: {len(result_indices)}")
    for x, y in zip(result_indices, result_samples):
        print(f"{x} {y}")
    visualize_discrete_signal(Signal_Frame4, result_indices,result_samples)

############################################################################################task5
#Use to test the Amplitude of DFT and IDFT
def SignalComapreAmplitude(SignalInput = [] ,SignalOutput= []):
    if len(SignalInput) != len(SignalOutput):
        return False
    else:
        for i in range(len(SignalInput)):
            if abs(SignalInput[i]-SignalOutput[i])>0.001:
                print("Wrong answer ! Test is failed")
                return False
        print(" Test is Successfully Passes")
        return True

def RoundPhaseShift(P):
    while P<0:
        p+=2*math.pi
    return float(P%(2*math.pi))

#Use to test the PhaseShift of DFT
def SignalComaprePhaseShift(SignalInput = [] ,SignalOutput= []):
    if len(SignalInput) != len(SignalOutput):
        return False
    else:
        for i in range(len(SignalInput)):
            A=round(SignalInput[i])
            B=round(SignalOutput[i])
            if abs(A-B)>0.0001:
                return False
        return True

########################


def Read_signal_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Skip the first two lines and get the number of samples from the third line
    num_samples = int(lines[2].strip())
    print("done1")
    # Initialize lists for amplitudes and phase shifts
    amplitudes = []
    phase_shifts = []
    
    # Process the subsequent lines for amplitude and phase shift values
    for i in range(3, 3 + num_samples):
        amplitude, phase_shift = map(lambda x: float(x.rstrip('f')), lines[i].strip().split())
        amplitudes.append(amplitude)
        phase_shifts.append(phase_shift)
    # print("lists are separated")
    # for amplitude, phase_shift in zip(amplitudes, phase_shifts):
    #     print("Amblitude : " ,amplitude, "    Phase : ", phase_shift)
    return amplitudes, phase_shifts

def IDFT(amplitudes, phase_shifts):

    sample_size=len(amplitudes)
    # Calculate x and y coordinates
    combined_coords_real = []  # real parts (x(k))
    combined_coords_imag = []  # imaginary parts (y(k))
    
    for amplitude, phase_shift in zip(amplitudes, phase_shifts):
        # real part: x = amplitude * cos(phase_shift)
        real_part = amplitude * np.cos(phase_shift)
        # imaginary part: y = amplitude * sin(phase_shift)
        imag_part = amplitude * np.sin(phase_shift)
        
        combined_coords_real.append(real_part)
        combined_coords_imag.append(imag_part)
    
    # Apply the DFT rule with Euler's formula for complex exponentials
    x_n = []
    for n in range(sample_size):
        sum_real = 0
        sum_imag = 0
        total_result=0
        for k in range(sample_size):
            # Calculate the complex exponential using Euler's formula:
            # e^(2*pi*j*k*n/N) = cos(2*pi*k*n/N) + j*sin(2*pi*k*n/N)
            angle = 2 * np.pi * k * n / sample_size
            exp_real = np.cos(angle)  # real part of e^(2*pi*j*k*n/N)
            exp_imag = np.sin(angle)  # imaginary part of e^(2*pi*j*k*n/N)
            # Multiply combined_coords[k] by the complex exponential (real and imaginary parts)
            sum_real += combined_coords_real[k] * exp_real - combined_coords_imag[k] * exp_imag
            sum_imag += combined_coords_real[k] * exp_imag + combined_coords_imag[k] * exp_real
            total_result=round(sum_real / sample_size)
        # Normalize by N and keep only the real part (imaginary part should be zero)
        # Since you want no imaginary part in the final x_n
        x_n.append(total_result)
    # for s in x_n:
    #     print("sample : ",s)
    
    return  x_n

def On_IDFT():
    # Read and parse the file
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    indices_test,amplitudes_test= Read_signal_file(r"d:\Habiba\7th semester\DSP\task 5\Lab7\Lab7\Test Cases\IDFT\Output_Signal_IDFT.txt")


    amplitudes, phase_shifts = Read_signal_file(file_path)
    n_values=IDFT(amplitudes, phase_shifts)
    
    size=len(n_values)
    K_values= list(range(size))

    res=SignalComapreAmplitude(n_values , amplitudes_test)
    print(res)

    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, K_values, n_values)
        messagebox.showinfo("Success", "Result saved successfully!")    

def DFT(samples ):
    sample_size=len(samples)
    dft_result = []

    for k in range(sample_size):
        sum_real = 0
        sum_imag = 0
        # Compute the k-th DFT component
        X_k = 0
        for n in range(sample_size):
            theta = -2 * np.pi * k * n / sample_size
            exp_real = np.cos(theta)  
            exp_imag = np.sin(theta) 
            sum_real += samples[n] * exp_real 
            sum_imag += samples[n] * exp_imag 
        X_k = complex(sum_real, sum_imag)
        dft_result.append(X_k)

    return dft_result

def visualize_result(fs ,dft_result):
    Magnitude=[]
    Phase=[]
    sample_size=len(dft_result)
    
    angular_frequency_step = (2 * np.pi * fs) / sample_size
    angular_frequencies = [k * angular_frequency_step for k in range(sample_size)]   
    for k, X_k in enumerate(dft_result):
        Magnitude.append(abs(X_k))
        Phase.append(np.angle(X_k))
        #print(f"Frequency bin {k}: {X_k} (Magnitude={abs(X_k):.5f}, Phase={np.angle(X_k):.5f})")
    

    # Create the figure for the plot
    fig, (ax,ax2) = plt.subplots(1,2,figsize=(12, 4))
    
    ax.stem(angular_frequencies, Magnitude)
    plt.title("Magnitude Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.grid()

    ax2.stem(angular_frequencies, Phase)
    plt.title("Phase Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Phase (radians)")
    plt.grid()
    
    # Embed the figure in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=Signal_Frame5)  
    canvas.draw()
    
    # Position the canvas in the window
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

    return Magnitude , Phase ,angular_frequencies

def On_DFT():
    # Read and parse the file
    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    

    amplitudes_test, phase_shifts_test = Read_signal_file(r"D:\Habiba\7th semester\DSP\task 5\Lab7\Lab7\Test Cases\DFT\output_DTF.txt")


    indices , samples  = ReadSignalFile(file_path)
    
    fs = simpledialog.askfloat("Input", "Enter sampling frequency :")
    
    if fs is None:  # If the user cancels the input
        messagebox.showinfo("Information", "No frequency entered.")
        return

    dft_result = DFT(samples)
    amplitudes,Phase , resindices =visualize_result(fs ,dft_result)
    
    res=SignalComapreAmplitude(amplitudes_test ,amplitudes)
    print(res)
    res2=SignalComaprePhaseShift(phase_shifts_test , Phase)
    print(res2)

    
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path,amplitudes,Phase)
        messagebox.showinfo("Success", "Result saved successfully!")    

############################################################################################task6
def Correation_Compare_Signals(file_name,Your_indices,Your_samples):      
    expected_indices=[]
    expected_samples=[]
    with open(file_name, 'r') as f:
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        while line:
            # process line
            L=line.strip()
            if len(L.split(' '))==2:
                L=line.split(' ')
                V1=int(L[0])
                V2=float(L[1])
                expected_indices.append(V1)
                expected_samples.append(V2)
                line = f.readline()
            else:
                break
    print("Current Output Test file is: ")
    print(file_name)
    print("\n")
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Shift_Fold_Signal Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Shift_Fold_Signal Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Correlation Test case failed, your signal have different values from the expected one") 
            return
    print("Correlation Test case passed successfully")
#######  1  ##########

def On_Direct_Correlation():

    file1 = filedialog.askopenfilename(title="Select First Signal File")
    if not file1:
        print("No file selected for Signal 1.")
        return

    indices1, samples1 = ReadSignalFile(file1)
    
    file2 = filedialog.askopenfilename(title="Select Second Signal File")
    if not file2:
        print("No file selected for Signal 2.")
        return

    indices2, samples2 = ReadSignalFile(file2)
    # frequency = frequency_entry.get()  # Get the frequency entered by the user
    result_indices,correlation=direct_correlation(samples1,samples2)
    Correation_Compare_Signals(r'D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\Correlation Task Files\Point1 Correlation\CorrOutput.txt',result_indices, correlation)
    plot_result(result_indices, correlation, Plot_Frame2)  # Assuming Plot_Frame2 is the frame for plotting


def direct_correlation(samples1, samples2):
    
    samples1 = np.array(samples1, dtype=float)
    samples2 = np.array(samples2, dtype=float)
    N = len(samples1)
    corr = []
    
    # Compute the normalization terms (magnitudes of samples1 and samples2)
    norm_samples1 = np.sqrt(np.sum(samples1 ** 2))
    norm_samples2 = np.sqrt(np.sum(samples2 ** 2))
    
    # Compute normalized cross-correlation for each lag
    for j in range(N):
        sum_corr = 0
        for n in range(N):
            sum_corr += samples1[n] * samples2[(n + j) % N]  # Circular indexing for lag
        corr.append(sum_corr / (norm_samples1 * norm_samples2))  # Normalize
    
    correlation = np.array(corr)
    result_indices = [i for i in range(N)]
    # print("result_indices:", result_indices)
    # print("Correlation:", correlation)
    return result_indices,correlation
######### 2  #######

def On_Correlation_time_delay():
    
    file1 = filedialog.askopenfilename(title="Select First Signal File")
    if not file1:
        print("No file selected for Signal 1.")
        return

    indices1, samples1 = ReadSignalFile(file1)
    
    file2 = filedialog.askopenfilename(title="Select Second Signal File")
    if not file2:
        print("No file selected for Signal 2.")
        return

    indices2, samples2 = ReadSignalFile(file2)
    result_indices,correlation=direct_correlation(samples1,samples2)
    plot_result(result_indices, correlation, Plot_Frame2)  # Assuming Plot_Frame2 is the frame for plotting
    frequency = frequency_entry.get()  # Get the frequency entered by the user
    try:
        frequency = float(frequency)  # Convert frequency to float
        print(f"Frequency entered: {frequency}")  # Example: Print frequency
        # Add your Direct Correlation logic here using the frequency
    except ValueError:
        print("Invalid frequency value entered!")
    max_lag = np.argmax(correlation)
    # Convert lag to time delay
    try:    
        time_delay = max_lag / frequency
        print(f"Time Delay (seconds): {time_delay}")
        return
    except ZeroDivisionError:
        print("Frequency cannot be zero.")
        return
   

# Triggering plot display (for testing)

######### 3  ###########
def On_Classify_signal():
    # Paths to folders containing Class A and Class B signals
    class_a_folder = r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\Correlation Task Files\point3 Files\Class 1"
    class_b_folder = r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\Correlation Task Files\point3 Files\Class 2"

    # Paths to Test signal
    test_file_path = r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\Correlation Task Files\point3 Files\Test Signals\Test1.txt"
    test_file_path2 = r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\Correlation Task Files\point3 Files\Test Signals\Test2.txt"
    
    # Load Test signals
    with open(test_file_path, 'r') as f:
        test_signal1 = np.array([int(line.strip()) for line in f])
    with open(test_file_path2, 'r') as f:
        test_signal2 = np.array([int(line.strip()) for line in f])

    # Classify Test signals
    result1 = classify_signal(test_signal1, class_a_folder, class_b_folder)
    print(f"The Test_1 signal belongs to: {result1}")
    result2 = classify_signal(test_signal2, class_a_folder, class_b_folder)
    print(f"The Test_2 signal belongs to: {result2}")

def load_signals_from_folder(folder_path):
    """Loads all signals from text files in a given folder."""
    signals = []
    for file_name in sorted(os.listdir(folder_path)):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path) and file_name.endswith('.txt'):
            with open(file_path, 'r') as f:
                signal = np.array([int(line.strip()) for line in f])
                signals.append(signal)
    return signals


def classify_signal(test_signal, class_a_folder, class_b_folder):
    """Classifies the test signal based on average maximum correlation."""
    # Load signals from Class A and Class B folders
    class_a_signals = load_signals_from_folder(class_a_folder)
    class_b_signals = load_signals_from_folder(class_b_folder)

    # Compute maximum correlations for Class A
    max_corr_a = []
    for sig in class_a_signals:
        _, corr = direct_correlation(test_signal, sig)
        max_corr_a.append(np.max(corr))  # Store maximum correlation value
    avg_max_corr_a = np.mean(max_corr_a)  # Average of maximum correlations
    print(f"Average Max Correlation (Class A): {avg_max_corr_a}")

    # Compute maximum correlations for Class B
    max_corr_b = []
    for sig in class_b_signals:
        _, corr = direct_correlation(test_signal, sig)
        max_corr_b.append(np.max(corr))  # Store maximum correlation value
    avg_max_corr_b = np.mean(max_corr_b)  # Average of maximum correlations
    print(f"Average Max Correlation (Class B): {avg_max_corr_b}")

    # Compare average maximum correlations
    if avg_max_corr_a > avg_max_corr_b:
        print("Classification Result: Class 1 (Down Movement)")
        return 'Class 1 Down Movement of signal'
    else:
        print("Classification Result: Class 2 (Up Movement)")
        return 'Class 2 Up Movement of signal'

# Example usage
############################################################################################task7


def Compare_Signals(file_name,Your_indices,Your_samples):      
    expected_indices=[]
    expected_samples=[]
    with open(file_name, 'r') as f:
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        while line:
            # process line
            L=line.strip()
            if len(L.split(' '))==2:
                L=line.split(' ')
                V1=int(L[0])
                V2=float(L[1])
                expected_indices.append(V1)
                expected_samples.append(V2)
                line = f.readline()
            else:
                break
    print("Current Output Test file is: ")
    print(file_name)
    print("\n")
    if (len(expected_samples)!=len(Your_samples)) and (len(expected_indices)!=len(Your_indices)):
        print("Test case failed, your signal have different length from the expected one")
        return
    for i in range(len(Your_indices)):
        if(Your_indices[i]!=expected_indices[i]):
            print("Test case failed, your signal have different indicies from the expected one") 
            return
    for i in range(len(expected_samples)):
        if abs(Your_samples[i] - expected_samples[i]) < 0.01:
            continue
        else:
            print("Test case failed, your signal have different values from the expected one") 
            return
    print("Test case passed successfully")

def On_Single(mode):

    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    

    input_window = tk.Toplevel()
    input_window.title("Enter Filter Parameters")
    
    # Create labels and entry widgets for each input
    tk.Label(input_window, text="Enter cut off frequency : ").grid(row=0, column=0)
    fc = tk.Entry(input_window)
    fc.grid(row=0, column=1)

    tk.Label(input_window, text="Enter Fs : ").grid(row=1, column=0)
    fs = tk.Entry(input_window)
    fs.grid(row=1, column=1)

    tk.Label(input_window, text="Enter Stop Band Attenuation : ").grid(row=2, column=0)
    Attenuation = tk.Entry(input_window)
    Attenuation.grid(row=2, column=1)

    tk.Label(input_window, text="Enter Transition Band : ").grid(row=3, column=0)
    Transition = tk.Entry(input_window)
    Transition.grid(row=3, column=1)
    


    def confirm():
                
        fc_value = int(fc.get())  # Convert entry to an integer
        fs_value = int(fs.get())
        Attenuation_value = int(Attenuation.get())
        Transition_value = int(Transition.get())
        print(fc_value)
        print(fs_value)
        
        if mode == "low":
            print("low-pass filter applied")
            Low(fc_value, fs_value, Attenuation_value,Transition_value,indices, values)
        
        elif mode == "high":
            print("high-stop filter applied")
            High(fc_value, fs_value, Attenuation_value,Transition_value,indices, values)
        

        


    # Button to confirm selection
    tk.Button(input_window, text="Confirm", command=confirm).grid(row=5, column=1)

def On_Band(mode):

    # Open a file dialog to select multiple signal files
    file_path = filedialog.askopenfilename(
        title="Select a Text File", 
        filetypes=(("Text Files", "*.txt"), ("All Files", "*.*"))
    )
    if not file_path:
        messagebox.showinfo("Information", "No files selected.")
        return
    
    # Read and add the signals
    indices, values = ReadSignalFile(file_path) 

    

    input_window = tk.Toplevel()
    input_window.title("Enter Filter Parameters")
    
    # Create labels and entry widgets for each input
    tk.Label(input_window, text="Enter cut off frequency 1 : ").grid(row=0, column=0)
    fc1 = tk.Entry(input_window)
    fc1.grid(row=0, column=1)

    tk.Label(input_window, text="Enter cut off frequency 2 : ").grid(row=1, column=0)
    fc2 = tk.Entry(input_window)
    fc2.grid(row=1, column=1)

    tk.Label(input_window, text="Enter Fs : ").grid(row=2, column=0)
    fs = tk.Entry(input_window)
    fs.grid(row=2, column=1)

    tk.Label(input_window, text="Enter Stop Band Attenuation : ").grid(row=3, column=0)
    Attenuation = tk.Entry(input_window)
    Attenuation.grid(row=3, column=1)

    tk.Label(input_window, text="Enter Transition Band : ").grid(row=4, column=0)
    Transition = tk.Entry(input_window)
    Transition.grid(row=4, column=1)
    


    def confirm():
                
        fc1_value = int(fc1.get())  # Convert entry to an integer
        fc2_value = int(fc2.get())
        fs_value = int(fs.get())
        Attenuation_value = int(Attenuation.get())
        Transition_value = int(Transition.get())
        if mode == "pass":
            print("Band-pass filter applied")
            Band_Pass(fc1_value, fc2_value ,fs_value, Attenuation_value,Transition_value,indices, values)
        
        elif mode == "stop":
            print("Band-stop filter applied")
            Band_Stop(fc1_value, fc2_value ,fs_value, Attenuation_value,Transition_value,indices, values)
        
        


    # Button to confirm selection
    tk.Button(input_window, text="Confirm", command=confirm).grid(row=6, column=1)

def Low(fc,fs,Attenuation,Transition,indices, values):
    print("before filter done")
    f_indices, f_values = DesignFIRFilter("low", fc, None, fs, Attenuation, Transition)
    print("before save")
    Compare_Signals(r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 1\LPFCoefficients.txt",f_indices, f_values)
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save filter coefficients", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path,f_indices, f_values)
        messagebox.showinfo("Success", "Result saved successfully!")
    compare_path=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 2\ecg_low_pass_filtered.txt"
    FilterAndPlot(f_indices, f_values, indices, values , fs,compare_path)
    # FilterinFrequencyDomain(f_indices, f_values, indices, values , fs,compare_path)

    return 

def High(fc,fs,Attenuation,Transition,indices, values):
    f_indices, f_values = DesignFIRFilter("high", fc, None, fs, Attenuation, Transition)
    Compare_Signals(r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 3\HPFCoefficients.txt",f_indices, f_values)
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save filter coefficients", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path,f_indices, f_values)
        messagebox.showinfo("Success", "Result saved successfully!")
    
    compare_path=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 4\ecg_high_pass_filtered.txt"
    FilterAndPlot(f_indices, f_values, indices, values , fs,compare_path)
    # FilterinFrequencyDomain(f_indices, f_values, indices, values , fs,compare_path)
    return 

def Band_Pass(fc1,fc2,fs,Attenuation,Transition,indices, values):
    f_indices, f_values = DesignFIRFilter("bandpass", fc1, fc2, fs, Attenuation, Transition)
    Compare_Signals(r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 5\BPFCoefficients.txt",f_indices, f_values)
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save filter coefficients", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path,f_indices, f_values)
        messagebox.showinfo("Success", "Result saved successfully!")
    compare_path=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 6\ecg_band_pass_filtered.txt"
    FilterAndPlot(f_indices, f_values, indices, values , fs,compare_path)
    # FilterinFrequencyDomain(f_indices, f_values, indices, values , fs,compare_path)
    return 

def Band_Stop(fc1,fc2,fs,Attenuation,Transition,indices, values):
    f_indices, f_values = DesignFIRFilter("bandstop", fc1, fc2, fs, Attenuation, Transition)
    Compare_Signals(r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 7\BSFCoefficients.txt",f_indices, f_values)
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save filter coefficients", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path, f_indices, f_values)
        messagebox.showinfo("Success", "Result saved successfully!")
    compare_path=r"D:\AAA\Level 4\Semester 7\DSP\Assignments\Correlation&Filter\FIR test cases\Testcase 8\ecg_band_stop_filtered.txt"
    FilterAndPlot(f_indices, f_values, indices, values , fs,compare_path)
    # FilterinFrequencyDomain(f_indices, f_values, indices, values , fs,compare_path)

    return 

def get_window(window_type, N):
    if window_type == "rectangular":
        # Rectangular window: All ones
        return np.ones(N)
    
    elif window_type == "hamming":
        # Hamming window: w(n) = 0.54 - 0.46 * cos(2 * pi * n / (N - 1))
        n = np.arange(N)
        return 0.54 - 0.46 * np.cos(2 * np.pi * n / (N-1))
    
    elif window_type == "hanning":
        # Hanning window: w(n) = 0.5 * (1 - cos(2 * pi * n / (N - 1)))
        n = np.arange(N)
        return 0.5 * (1 - np.cos(2 * np.pi * n / (N-1)))
    
    elif window_type == "blackman":
    # Blackman window: w(n) = 0.42 - 0.5 * cos(2 * pi * n / (N - 1)) + 0.08 * cos(4 * pi * n / (N - 1))
        n = np.arange(N)
        return 0.42 - 0.5 * np.cos(2 * np.pi * n / (N - 1)) + 0.08 * np.cos(4 * np.pi * n / (N - 1))
    else:
        raise ValueError("Unsupported window type")

def DesignFIRFilter(filter_type, fc1, fc2, fs, Attenuation, Transition):

    # Compute the filter coefficients
    if filter_type == "low":
        fc1=(fc1+(Transition/2))/fs
    elif filter_type == "high":
        fc1=(fc1-(Transition/2))/fs
    elif filter_type == "bandpass":
        fc1=(fc1-(Transition/2))/fs
        fc2=(fc2+(Transition/2))/fs
    elif filter_type == "bandstop":
        fc1=(fc1+(Transition/2))/fs
        fc2=(fc2-(Transition/2))/fs
    else:
        raise ValueError("Invalid filter type")

    
    # Select the window and compute N
    if Attenuation <= 21:
        window = "rectangular"  # Rectangular window
        N = (0.9 * fs)/Transition
    elif Attenuation <= 44:
        window = "hanning"  # Hanning window
        N = (3.1 * fs)/Transition
    elif Attenuation <= 53:
        window = "hamming"  # Hamming window
        N = (3.3 * fs)/Transition
    elif Attenuation <= 74:
        window = "blackman"  # Blackman window
        N = (5.5 * fs)/Transition
    else:
        raise ValueError("Unsupported attenuation for the available windows (Rectangular: <=21 dB, Hamming: <=53 dB)")
    
    # Round N up to the next odd integer
    N = math.ceil(N)  # Round up to the next integer
    if N % 2 == 0:    # If even, make it odd
        N += 1
    print("afer N step")

    # Define range for n
    indices = np.arange(-(N-1)//2, (N-1)//2 + 1)

    # Compute the ideal impulse response h_d(n)
    h_d = np.zeros_like(indices, dtype=float)
    if filter_type == "low":
        for i, n in enumerate(indices):
            if n == 0:
                h_d[i] = 2 * fc1
            else:
                h_d[i] = np.sin(2 * np.pi * fc1 * n) / (np.pi * n)
    elif filter_type == "high":
        for i, n in enumerate(indices):
            if n == 0:
                h_d[i] = 1 - 2 * fc1
            else:
                h_d[i] = -np.sin(2 * np.pi * fc1 * n) / (np.pi * n)
    elif filter_type in ["bandpass", "bandstop"]:
        for i, n in enumerate(indices):
            if n == 0:
                h_d[i] = 2 * (fc2 - fc1) if filter_type == "bandpass" else 1 - 2 * (fc2 - fc1)
            else:
                h_d[i] = (
                    (np.sin(2 * np.pi * fc2 * n) - np.sin(2 * np.pi * fc1 * n)) / (np.pi * n)
                )
                if filter_type == "bandstop":
                    h_d[i] = -h_d[i]    

    print("afer LOOP step")
     # Compute w(n) based on the selected window
    w = get_window(window, N)
    print("after W step")

    # Compute h_d(n)
    values = h_d * w
    print("after * step")

    return indices, values

def FilterAndPlot(f_indices, f_values, indices, values , fs,compare_path):
    # Time-domain convolution
    result_indices, result_samples = manual_convolution_with_signals(indices, values, f_indices, f_values)
    Compare_Signals(compare_path,result_indices, result_samples)
    
    plot_result(result_indices, result_samples, Plot_Frame)
    save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Correlation of signal and filter", filetypes=[("Text Files", "*.txt")])
    if save_path:
        save_signal_to_file(save_path,result_indices, result_samples,)
        messagebox.showinfo("Success", "Result saved successfully!")
    return 
    #frequency domain
    
def FilterinFrequencyDomain(f_indices, f_values, indices, values , fs,compare_path):
    
    Min_size = min(len(f_values), len(values))
    
    # f_values_padded = np.pad(f_values, (0, max_size - len(f_values)), mode='constant')
    # values_padded = np.pad(values, (0, max_size - len(values)), mode='constant')
    
    # Perform DFT on the zero-padded filter and signal
    f=f_values[:Min_size]
    s=values[:Min_size]
    FilterAfterDft = DFT(f)
    SignalAfterDft = DFT(s)
    resulted = np.conj(FilterAfterDft)* SignalAfterDft
    Magnitude=[]
    Phase=[]
    sample_size=len(resulted)
    
    angular_frequency_step = (2 * np.pi * fs) / sample_size
    angular_frequencies = [k * angular_frequency_step for k in range(sample_size)]   
    for k, X_k in enumerate(resulted):
        Magnitude.append(abs(X_k))
        Phase.append(np.angle(X_k))
        
     
    
    resultAfterIDFT =IDFT(Magnitude,Phase)
    result_indices = np.linspace(0, len(resultAfterIDFT) / fs, len(resultAfterIDFT), endpoint=False)
    # print(result_indices)
    print("values##################")   
    print(resultAfterIDFT)
    Compare_Signals(compare_path,result_indices, resultAfterIDFT)

    # Save the filtered signal to a file
    # save_path = filedialog.asksaveasfilename(defaultextension=".txt", title="Save Result Signal", filetypes=[("Text Files", "*.txt")])
    # if save_path:
    #     save_signal_to_file(save_path, result_indices, result_values)
    #     messagebox.showinfo("Success", "Result saved successfully!")     
    #         Plot the original and filtered signals
            

    # plt.figure(figsize=(10, 6))

    # # Plot the original signal
    # plt.subplot(2, 1, 1)
    # plt.plot(indices, values, label="Original Signal")
    # plt.title("Original Signal in Time Domain")
    # plt.xlabel("Time (s)")
    # plt.ylabel("Amplitude")
    # plt.grid(True)
    # plt.legend()

    # # Plot the filtered signal (after IDFT)
    # plt.subplot(2, 1, 2)
    # plt.plot(indices, resultAfterIDFT, label="Filtered Signal (After IDFT)", color="orange")
    # plt.title("Filtered Signal in Time Domain (After IDFT)")
    # plt.xlabel("Time (s)")
    # plt.ylabel("Amplitude")
    # plt.grid(True)
    # plt.legend()

    # # Show the plots
    # plt.tight_layout()
    # plt.show()
    
    return resultAfterIDFT
    
# Define the main function
def plot_result(result_indices, result_samples, plot_frame):
    # If there is an existing plot, remove it
    for widget in plot_frame.winfo_children():
        widget.destroy()  # Remove the old plot

    # Create a new figure and axis for plotting
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot the result data
    ax.plot(result_indices, result_samples, label='Filtered Signal', color='b')
    
    # Set labels and title
    ax.set_xlabel('Index')
    ax.set_ylabel('Amplitude')
    ax.set_title('Filtered Signal')
    ax.legend()

    # Embed the plot in the Tkinter GUI
    new_canvas = FigureCanvasTkAgg(fig, master=plot_frame)  # Use the Plot_Frame here
    new_canvas.draw()  # Render the plot
    new_canvas.get_tk_widget().pack()  # Pack it into the Plot_Frame

    return   # Return the new canvas for future clearing

    
    
root = tk.Tk()
root.title("Navbar with Multiple Tabs")
root.geometry("2000x800")

style = ttk.Style()
style.configure('TNotebook.Tab', padding=[20, 10])  # Increase padding to make tabs wider

# Create a Notebook widget, which will act as the navbar with tabs
notebook = ttk.Notebook(root, style='TNotebook')

# Define different frames for each tab (these will contain different content)
frame1 = ttk.Frame(notebook)
frame2 = ttk.Frame(notebook)
frame3 = ttk.Frame(notebook)
frame4 = ttk.Frame(notebook)
frame5 = ttk.Frame(notebook)
frame6 = ttk.Frame(notebook)
frame7 = ttk.Frame(notebook)
# Add frames (as tabs) to the notebook with tab titles
notebook.add(frame1, text="Task 1" )
notebook.add(frame2, text="Task 2" )
notebook.add(frame3, text="Task 3" )
notebook.add(frame4, text="Task 4" )
notebook.add(frame5, text="Task 5" )
notebook.add(frame6, text="Task 6" )
notebook.add(frame7, text="Task 7" )

# Pack the notebook widget so it fills the window
notebook.pack(expand=True, fill="both")
# Center the frame vertically and horizontally

label1 = tk.Label(frame1, text=" Task 1", font=("Arial", 14) ,fg="Green")
label1.pack(pady=10)
#######################
button_frame = tk.Frame(frame1)
button_frame.pack(pady=10, anchor="center") 
Signal_Frame = tk.Frame(frame1)
Signal_Frame.pack(pady=10, anchor="center") 
button2 = tk.Button(button_frame, text="Display Signal",fg="white", bg="green" ,font=("Arial", 12) ,command=on_display_click)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame, text="Add Signals",fg="white", bg="green",font=("Arial", 12) ,command=on_add_signals_click)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame, text="Subtract Signals" ,fg="white", bg="green",font=("Arial", 12) ,command=on_sub_signals_click)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame, text="Multiply by constant" ,fg="white", bg="green",font=("Arial", 12) ,command=On_Multiply_Click)
button2.pack(side="left",padx=5)
button2 = tk.Button(button_frame, text="Reverse of a signal",fg="white", bg="green" ,font=("Arial", 12),command=on_rev_signals_click)
button2.pack(side="left",padx=5)
button2 = tk.Button(button_frame, text="Shifting a signal",fg="white", bg="green" ,font=("Arial", 12),command=on_shift_signal_click)
button2.pack(side="left",padx=5)

# Content for Tab 2
label2 = tk.Label(frame2, text="Task 2", font=("Arial", 14))
label2.pack(pady=20)
#########################

button_frame2 = tk.Frame(frame2)
button_frame2.pack(pady=10, anchor="center") 
Signal_Frame2 = tk.Frame(frame2)
Signal_Frame2.pack(pady=10, anchor="center") 
button2 = tk.Button(button_frame2, text="Display Discrete Signal",fg="white", bg="green" ,font=("Arial", 12) ,command=on_display_discrete)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame2, text="Display Continuos Signal",fg="white", bg="green",font=("Arial", 12) ,command=on_display_continuos)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame2, text="Cos Signal" ,fg="white", bg="green",font=("Arial", 12) ,command=on_cos_signal_click)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame2, text="Sin Signal" ,fg="white", bg="green",font=("Arial", 12) ,command=on_sin_signal_click)
button2.pack(side="left",padx=5)


# Content for Tab 3
label3 = tk.Label(frame3, text="Task 3", font=("Arial", 14))
label3.pack(pady=20)

button_frame3 = tk.Frame(frame3)
button_frame3.pack(pady=10, anchor="center") 
Signal_Frame3 = tk.Frame(frame3)
Signal_Frame3.pack(pady=10, anchor="center") 
button2 = tk.Button(button_frame3, text="Quantization",fg="white", bg="green" ,font=("Arial", 12) ,command=on_quant_signal)
button2.pack(side="left", padx=5)



# Content for Tab 4
label4 = tk.Label(frame4, text="Task 4", font=("Arial", 14))
label4.pack(pady=20)

button_frame4 = tk.Frame(frame4)
button_frame4.pack(pady=10, anchor="center") 
Signal_Frame4 = tk.Frame(frame4)
Signal_Frame4.pack(pady=10, anchor="center") 
button2 = tk.Button(button_frame4, text="Moving",fg="white", bg="green" ,font=("Arial", 12) ,command=on_move_signal)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame4, text=" First Derivative",fg="white", bg="green" ,font=("Arial", 12) ,command=on_first_D)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame4, text=" second Derivative",fg="white", bg="green" ,font=("Arial", 12) ,command=on_second_D)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame4, text="Convolution",fg="white", bg="green" ,font=("Arial", 12) ,command= on_conv_signal)
button2.pack(side="left", padx=5)


# Content for Tab 5
label5 = tk.Label(frame5, text="Task 5", font=("Arial", 14))
label5.pack(pady=20)

button_frame5 = tk.Frame(frame5)
button_frame5.pack(pady=10, anchor="center") 
Signal_Frame5 = tk.Frame(frame5)
Signal_Frame5.pack(pady=10, anchor="center") 
button2 = tk.Button(button_frame5, text="Apply IDFT",fg="white", bg="green" ,font=("Arial", 12),width=10, height=2 ,command=On_IDFT)
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame5, text="Apply DFT",fg="white", bg="green" ,font=("Arial", 12),width=10, height=2 ,command=On_DFT)
button2.pack(side="left", padx=5)
############################################# Content for Tab 6#############################################################
# Label for Task Correlation
label6 = tk.Label(frame6, text="Task Correlation", font=("Arial", 16))
label6.pack(pady=10)

# Button and Frequency Input Frame
button_frame6 = tk.Frame(frame6)
button_frame6.pack(pady=20)

# Button 1: Direct Correlation, Button 2: Time Delay, Button 3: Classify Signal
button1 = tk.Button(button_frame6, text="Direct Correlation", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=On_Direct_Correlation)
button1.grid(row=0, column=0, pady=5, padx=5)

button2 = tk.Button(button_frame6, text="Time Delay", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=On_Correlation_time_delay)
button2.grid(row=0, column=1, pady=5, padx=5)

button3 = tk.Button(button_frame6, text="Classify A or B", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=On_Classify_signal)
button3.grid(row=0, column=2, pady=5, padx=5)

# Frame for Frequency Input (below the buttons)
input_frame6 = tk.Frame(frame6)
input_frame6.pack(pady=5)

# Label and Entry for Frequency
frequency_label = tk.Label(input_frame6, text="Frequency (Hz):", font=("Arial", 12))
frequency_label.pack(side="left", padx=5)

frequency_entry = tk.Entry(input_frame6, font=("Arial", 12), width=10)
frequency_entry.pack(side="left", padx=5)

# Frame for Plotting (below the frequency input)
Plot_Frame2 = tk.Frame(frame6)
Plot_Frame2.pack(pady=10)


############################################


# Create a label for Task 7
label7 = tk.Label(frame7, text="Task 7", font=("Arial", 14))
label7.pack(pady=20)

# Create the button frame
button_frame7 = tk.Frame(frame7)
button_frame7.pack(pady=10, anchor="center")

# Create the frame for buttons
Signal_Frame7 = tk.Frame(frame7)
Signal_Frame7.pack(pady=10, anchor="center")

# Create buttons for different filters
button2 = tk.Button(button_frame7, text="Apply low-pass", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=lambda: On_Single("low"))
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame7, text="Apply high-pass", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=lambda: On_Single("high"))
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame7, text="Apply band-pass", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=lambda: On_Band("pass"))
button2.pack(side="left", padx=5)
button2 = tk.Button(button_frame7, text="Apply band-stop", fg="white", bg="green", font=("Arial", 12), width=15, height=2, command=lambda: On_Band("stop"))
button2.pack(side="left", padx=5)

# Create a new frame for the plot
Plot_Frame = tk.Frame(frame7)
Plot_Frame.pack(pady=20, anchor="center")



# Start the main application loop
root.mainloop()    
```
