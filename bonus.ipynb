```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from tkinter import filedialog, messagebox,simpledialog
import math

def direct_correlation(samples1, samples2):
    
    samples1 = np.array(samples1, dtype=float)
    samples2 = np.array(samples2, dtype=float)
    N = len(samples1)
    corr = []
    
    norm_samples1 = np.sqrt(np.sum(samples1 ** 2))
    norm_samples2 = np.sqrt(np.sum(samples2 ** 2))
    
    for j in range(N):
        sum_corr = 0
        for n in range(N):
            sum_corr += samples1[n] * samples2[(n + j) % N]  
        corr.append(sum_corr / (norm_samples1 * norm_samples2)) 
    
    correlation = np.array(corr)
    result_indices = [i for i in range(N)]
    # print("result_indices:", result_indices)
    # print("Correlation:", correlation.shape)
    return result_indices,correlation
# Function to generate a sine wave signal
def generate_sine_signal(amplitude, frequency, num_samples):
    t = np.linspace(0, 1, int(num_samples))
    sine_signal = float(amplitude) * np.sin(2 * np.pi * float(frequency) * t)
    return sine_signal
# Function to plot the sine wave signal
def plot_sine_wave(amplitude_entry, frequency_entry, samples_entry, canvas_frame):
    amplitude = float(amplitude_entry.get())
    frequency = float(frequency_entry.get())
    num_samples = int(samples_entry.get())

    sine_signal = generate_sine_signal(amplitude, frequency, num_samples)

    for widget in canvas_frame.winfo_children():
        widget.destroy()

    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(np.linspace(0, 1, num_samples), sine_signal)
    ax.set_title("Sine Wave Signal", fontsize=18, fontweight="bold")
    ax.set_xlabel("Time (s)", fontsize=14)
    ax.set_ylabel("Amplitude", fontsize=14)

    # Display the plot in the Tkinter canvas frame
    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

def sin_autocorelation(amplitude_entry, frequency_entry, samples_entry, canvas_frame):
    amplitude = float(amplitude_entry.get())
    frequency = float(frequency_entry.get())
    num_samples = int(samples_entry.get())

    sine_signal = generate_sine_signal(amplitude, frequency, num_samples)
    index,autocorr_sine = direct_correlation(sine_signal, sine_signal)

    for widget in canvas_frame.winfo_children():
        widget.destroy()

    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(np.array(index) - (len(sine_signal) / 2), autocorr_sine)
    ax.set_title("Auto-Correlation of Sine Wave Signal", fontsize=18, fontweight="bold")
    ax.set_xlabel("Lag", fontsize=14)
    ax.set_ylabel("Correlation", fontsize=14)

    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

def Gen_AWGN_Signal(samples_entry, canvas_frame):
    num_samples = int(samples_entry.get())

    noise_signal = np.random.normal(0, 1, num_samples)
    
    for widget in canvas_frame.winfo_children():
        widget.destroy()

    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(np.linspace(0, 1, num_samples), noise_signal)
    ax.set_title("Noise Signal", fontsize=18, fontweight="bold")
    ax.set_xlabel("Time (s)", fontsize=14)
    ax.set_ylabel("Amplitude", fontsize=14)

    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
    return noise_signal

def AddSine_Cos(amplitude_entry, frequency_entry, samples_entry, canvas_frame):
    amplitude = float(amplitude_entry.get())
    frequency = float(frequency_entry.get())
    num_samples = int(samples_entry.get())

    sine_signal = generate_sine_signal(amplitude, frequency, num_samples)
    noise_signal = np.random.normal(0, 1, num_samples)
    corrupted_signal = sine_signal + noise_signal
    
    for widget in canvas_frame.winfo_children():
        widget.destroy()
        
    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(np.linspace(0, 1, num_samples), corrupted_signal)
    ax.set_title("Corrupted Signal (Sine Wave + Noise)", fontsize=18, fontweight="bold")
    ax.set_xlabel("Time (s)", fontsize=14)
    ax.set_ylabel("Amplitude", fontsize=14)
    

    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
    return corrupted_signal

def Addithion_autocorrelation(amplitude_entry, frequency_entry, samples_entry, canvas_frame):
    amplitude = amplitude_entry
    frequency = frequency_entry
    num_samples = samples_entry

    corrupted_signal = AddSine_Cos(amplitude, frequency, num_samples, canvas_frame)
    index,autocorr_corrupted = direct_correlation(corrupted_signal, corrupted_signal)

    for widget in canvas_frame.winfo_children():
        widget.destroy()

    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(np.array(index) - (len(corrupted_signal) / 2), autocorr_corrupted)
    ax.set_title("Auto-Correlation of Corrupted Signal", fontsize=18, fontweight="bold")
    ax.set_xlabel("Lag", fontsize=14)
    ax.set_ylabel("Correlation", fontsize=14)

    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
    
# Function to create the GUI
def create_gui():
    # Create the main window
    root = tk.Tk()
    root.title("Periodic Signal Detection")
    root.geometry("1200x600")  # Set the window size
    root.configure(bg="#f0f0f0")  # Set the background color

    # Create a custom style for the GUI
    style = ttk.Style(root)
    style.theme_use("clam")
    style.configure("TLabel", foreground="#333333", font=("Helvetica", 12))
    style.configure("TEntry", fieldbackground="#ffffff", foreground="#333333", font=("Helvetica", 12))
    style.configure("TButton", background="#af504c", foreground="#ffffff", font=("Helvetica", 12), padding=10)

    # Create input fields and buttons
    input_frame = ttk.Frame(root, padding=20)
    input_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

    amplitude_label = ttk.Label(input_frame, text="Amplitude:")
    amplitude_label.pack(pady=1)
    amplitude_entry = ttk.Entry(input_frame)
    amplitude_entry.pack(pady=5)

    frequency_label = ttk.Label(input_frame, text="Frequency (Hz):")
    frequency_label.pack(pady=1)
    frequency_entry = ttk.Entry(input_frame)
    frequency_entry.pack(pady=5)

    samples_label = ttk.Label(input_frame, text="Number of Samples:")
    samples_label.pack(pady=1)
    samples_entry = ttk.Entry(input_frame)
    samples_entry.pack(pady=5)

    button_frame = ttk.Frame(input_frame, padding=10)
    button_frame.pack(pady=50)

    sine_wave_button = ttk.Button(button_frame, text="Sine Wave", command=lambda: plot_sine_wave(amplitude_entry, frequency_entry, samples_entry, canvas_frame),width=20)
    sine_wave_button.pack(side=tk.TOP, padx=10, pady=5)

    autocorr_button = ttk.Button(button_frame, text="Auto-Correlation", command=lambda: sin_autocorelation(amplitude_entry, frequency_entry, samples_entry, canvas_frame),width=20)
    autocorr_button.pack(side=tk.TOP, padx=10, pady=5)

    fft_button = ttk.Button(button_frame, text="Generate_AWGN_Signal", command=lambda: Gen_AWGN_Signal(samples_entry, canvas_frame),width=20)
    fft_button.pack(side=tk.TOP, padx=10, pady=5)

    psd_button = ttk.Button(button_frame, text="Addition_Sin_Cos", command=lambda: AddSine_Cos(amplitude_entry, frequency_entry, samples_entry, canvas_frame),width=20)
    psd_button.pack(side=tk.TOP, padx=10, pady=5)

    psd_button = ttk.Button(button_frame, text="Sin_cos_AutoCorrelation", command=lambda: Addithion_autocorrelation(amplitude_entry, frequency_entry, samples_entry, canvas_frame),width=20)
    psd_button.pack(side=tk.TOP, padx=10, pady=5)

    # Create a frame for the Matplotlib canvas
    canvas_frame = ttk.Frame(root, width=600, height=600, borderwidth=2, relief="solid")
    canvas_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    # Run the main loop
    root.mainloop()

# Run the GUI
create_gui()
```
